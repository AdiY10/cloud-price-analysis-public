```python
import json as json
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
import requests

```

#### Access the API end point for 'Vitual Machine' and create JSON file ../data/raw/

```python
r = requests.get('https://prices.azure.com/api/retail/prices?$filter=serviceName eq \'Virtual Machines\'')
with open('../../data/raw/azure_vm.json','w') as fd:
    fd.write(r.text)


```

```python
df = pd.read_json("../../data/raw/azure_vm.json")
df.head()
```

#### Understand shape of data

```python
print("Shape of the data\n\n", df.shape)
print("\n\n")
print("Is null?\n\n", df.isnull().sum())
print("\n\n")
print("Columnwise Datatype\n\n", df.dtypes)
df.describe()
```

```python
#### Split data into multiple components, create series out of dataframes, List "Items" is carrying actual data items
```

```python
df_items = df["Items"]
df_items = df_items.fillna("NA")
type(df_items)
dfi_l=len(df_items)
dfi_l
```

```python
#print dict
df_items
```

#### Normalize data into tabular format

```python
dfi_normalize = pd.json_normalize(df_items)
(dfi_normalize).fillna("NA")
dfi_normalize
```

#### Understand unique Product Names from Items

```python
pd.unique(dfi_normalize["skuName"])
#pd.Series((dfi_s["productName"])).unique()
#pd.Series((dfi_normalize["armRegionName"])).unique()
```

```python
dfi_s=dfi_normalize[["armRegionName","productName","skuName","retailPrice","currencyCode","serviceName","serviceFamily"]]
#dfi_s=dfi_normalize[["armRegionName","productName","skuName","retailPrice"]]
dfi_s
```

#### Group by columns ('armRegionName','productName','skuName','serviceName','serviceFamily') to get region wise Max Price and Min Price

```python
#data = dfi_s.groupby('productName').transform(lambda x: (x - x.min()) / x.max()- x.min())
#grouper = dfi_s.groupby(['armRegionName','productName','skuName'])['retailPrice'].agg(['max','min'])
result = dfi_s.groupby(['armRegionName','productName','skuName','serviceName','serviceFamily']).agg(
    maxPrice=('retailPrice','max'),minPrice=('retailPrice','min'))

result
```
